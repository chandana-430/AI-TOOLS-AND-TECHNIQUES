# Linear graph

```python
from matplotlib import style #to add style
```

```python
x=np.array([1,2,3])
y=np.array([5,7,4])
plt.plot(x,y,label="line")
plt.title("sample linear graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
print(plt.style.available)
```

```python
x1=np.array([4,8,10])
y1=np.array([13,6,6])
x2=np.array([6,9,10])
y2=np.array([6,13,7])
plt.plot(x1,y1,label="line one",color='m',linewidth=6)
plt.plot(x2,y2,label="line two",color="g",linewidth=6)
plt.title("sample linear graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid(True,color='k')
style.use("_classic_test_patch")
plt.show()
```

# Bar graph

```python
x1=np.array([4,8,10])
y1=np.array([13,6,6])
x2=np.array([6,9,10])
y2=np.array([6,13,7])
plt.bar(x1,y1,label="bar one",color='m')
plt.bar(x2,y2,label="bar two",color='c')
plt.legend()
plt.xlabel("bar number")
plt.ylabel("bar values")
plt.title("sample bar graph")
plt.show()
```

# Histogram
* numerical data/quantitative

```python
x=[5,10,15,20,25,26,28,40,45,28,11,58,85]
y=[0,10,20,30,40,50,60,70,90,100] #increasing order
plt.hist(x,y,histtype="bar",label="hist",rwidth=0.5) #syntax important
plt.xlabel("x hist")
plt.ylabel("y hist")
plt.title("histogram")
plt.legend()
plt.show()
```

# Scatter plot

```python
x1=np.array([4,8,10])
y1=np.array([13,6,6])
plt.scatter(x1,y1,label="scat",color='r')
plt.xlabel("x scat")
plt.ylabel("y scat")
plt.title("scatterplot")
plt.legend()
plt.show()
```

# Stack plot/area plot

```python
a=np.array([1,2,3,4,5])
b=np.array([2,3,4,3,2])
c=np.array([7,6,5,8,9])
d=np.array([8,7,5,4,10])
plt.plot([],color='g',label="green",linewidth=5)
plt.plot([],color='m',label='pink',linewidth=5)
plt.plot([],color='b',label='blue',linewidth=5)
plt.plot([],color='r',label='red',linewidth=5)
plt.stackplot(a,b,c,d,colors=['g','m','b','r']) #important
plt.xlabel("x stack")
plt.ylabel("y stack")
plt.title("stack_plot")
plt.legend()
plt.show()
```

# Pie chart

```python
a=np.array([9,2,3,10])
b=np.array(['red','blue','green','black'])
colors=['r','b','g','k']
plt.pie(a,labels=b,colors=colors,startangle=90,shadow=True) #syntax
plt.title("Pie_chart")
plt.show()
```
